```python
import pandas as pd

temps= [ [ 22.99, 20.6, 24.23, 23.51,1009,83,19.95,"2023-07-25"],
         [ 23.99, 21.6, 22.23, 23.51,1010,83,18.95,"2023-07-24"]]

cols=[   "temp",
    "temp_min",
    "temp_max",
    "feels_like",
    "pressure",
    "humidity",
    "dew_point",
    "date"
]


summer=pd.DataFrame(temps,columns=cols)

summer.set_index(['date'],inplace=True)

summer
```

```python
summer.mean()
```
